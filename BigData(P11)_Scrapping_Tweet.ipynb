```python
import pandas as pd
# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"
# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=",")
# Display the DataFrame
display(df)
```

```python
# Cek jumlah data yang didapatkan
num_tweets = len(df)
print(f"Jumlah tweet dalam dataframe adalah {num_tweets}.")
```

```python
import copy
data = copy.deepcopy(df[['full_text']])
data["full_text"]= df["full_text"].astype(str)
data['full_text'].head(10)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, udf, regexp_replace
from pyspark.sql.types import StringType
import re

# Membuat SparkSession
spark = SparkSession.builder \
    .appName("Data Cleaning") \
    .getOrCreate()

# Membuat DataFrame PySpark
columns = ["full_text"]
df = spark.createDataFrame(data, columns)

# Fungsi untuk membersihkan teks
def clean_text(text):
    # Lowercase
    text = text.lower()
    # Menghapus mentions (@username) dan hashtags (#hashtag)
    text = re.sub(r"(@[A-Za-z0-9_]+)|(#\w+)", "", text)
    # Menghapus URL
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    # Menghapus angka
    text = re.sub(r"\d+", "", text)
    # Menghapus karakter khusus dan tanda baca
    text = re.sub(r"[^\w\s]", "", text)
    # Menghapus spasi berlebihan
    text = re.sub(r"\s+", " ", text).strip()
    return text

# Mendefinisikan UDF PySpark untuk membersihkan teks
clean_text_udf = udf(clean_text, StringType())

# Menggunakan UDF untuk membersihkan kolom 'full_text'
cleaned_df = df.withColumn("cleaned_text", clean_text_udf(col("full_text")))

# Menampilkan hasil data yang telah dibersihkan (hanya 10 baris pertama)
cleaned_df.select("cleaned_text").show(10, truncate=False)
```

```python
# Menyimpan DataFrame yang telah dibersihkan ke file CSV
cleaned_df.select("cleaned_text").write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv("/content/tweets-data/full_text")
```

```python
from pymongo import MongoClient
import pandas as pd
import glob
import os

# Baca file CSV menjadi Pandas DataFrame
file_path = "/content/tweets-data/full_text/*.csv"
all_files = glob.glob(file_path)
df_list = [pd.read_csv(filename) for filename in all_files]
cleaned_df_pandas = pd.concat(df_list, ignore_index=True)

# Konversi DataFrame menjadi list of dictionaries
data_hasil_etl = cleaned_df_pandas.to_dict(orient="records")

# Membuat koneksi ke MongoDB Atlas
# Ganti dengan MongoDB connection string Anda sendiri
# Format: mongodb+srv://<username>:<password>@<cluster>.mongodb.net/
mongo_uri = os.getenv('MONGODB_URI')

if not mongo_uri:
    raise ValueError("MONGODB_URI not found in environment variables!")

client = MongoClient(mongo_uri)

# Pilih database dan koleksi
db = client['hasil_etl']  # nama database
collection = db['text_clean']  # nama collection

# Menyimpan data ke MongoDB
try:
    result = collection.insert_many(data_hasil_etl)
    print(f"✅ Berhasil menyimpan {len(result.inserted_ids)} dokumen ke MongoDB")
except Exception as e:
    print(f"❌ Terjadi kesalahan: {e}")
```
